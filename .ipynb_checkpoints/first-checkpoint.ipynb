# Load Data

```python
import pandas as pd
```

```python
data = pd.read_csv('18K cleaned Data analyst jobs.csv')
```

```python
#Strips any leading or trailing spaces from string values, while leaving other values unchanged

cleaned_data = data.applymap(lambda x: x.strip() if isinstance(x, str) else x)
cleaned_data.to_csv('dataset_ver2.csv', index=False)
```

```python
data = pd.read_csv('dataset_ver2.csv')
```

```python
# Display the shape of the dataframe
print(f'The dataset has {data.shape[0]} rows and {data.shape[1]} columns.')

# Display the column names
print(f'Column names: {data.columns.tolist()}\n')

# Display the data types of each column
print(f'Data types:\n{data.dtypes}\n')

# Check for missing values
print('Missing values:')
print(data.isnull().sum())
```

```python
top_titles = data['title'].value_counts().head(10)
print(top_titles)
```

```python
top_locations = data['location'].value_counts().head(10)
print(top_locations)
```

```python
top_vias = data['via'].value_counts().head(10)
print(top_vias)
```

```python
first_job_description = data.loc[0, 'description']
print(first_job_description)
```

```python
import matplotlib.pyplot as plt

top_titles = data['title'].value_counts().head(10)

plt.figure(figsize=(10, 6))
top_titles.plot(kind='bar')
plt.title('Top 10 Job Titles')
plt.xlabel('Job Titles')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
