```python
import spacy
from spacy.matcher import PhraseMatcher
from skillNer.general_params import SKILL_DB
from skillNer.skill_extractor_class import SkillExtractor
import pandas as pd
from tqdm import tqdm

# Load the spacy model
nlp = spacy.load("en_core_web_lg")

# Initialize the skill extractor
skill_extractor = SkillExtractor(nlp, SKILL_DB, PhraseMatcher)

# Function to extract skills and return a list
def extract_skills(description):
    annotations = skill_extractor.annotate(description)
    full_matches = annotations['results']['full_matches']
    ngram_scored = annotations['results']['ngram_scored']
    
    extracted_skills = [match['doc_node_value'] for match in full_matches]
    extracted_skills += [match['doc_node_value'] for match in ngram_scored]
    
    return extracted_skills

df = pd.read_csv('1.7_dataset.csv')


des = df['description'][0]

output = extract_skills(des)
print(output)

```

```python
# Load your CSV dataset
df = pd.read_csv('1.7_dataset.csv')

# Create an empty list to store extracted skills
extracted_skills_list = []

# Iterate through each row in the dataset and extract skills
for index, row in tqdm(df.iterrows(), total=len(df)):
    description = row['description']
    extracted_skills = extract_skills(description)
    extracted_skills_list.append(extracted_skills)

# Add the extracted skills list as a new column to the dataframe
df['extracted skills'] = extracted_skills_list
```

```python
# Save the updated DataFrame with the 'extracted_skills' column
output_path = '1.8_dataset.csv'
df.to_csv(output_path, index=False)
```
